##Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
Ans -
- Bokeh, glyphs are visual representations of data points that are used to create plots. They are graphical elements such as circles, lines, rectangles, bars, or patches that represent individual data points or groups of data points in a plot.

- Glyphs can be added to a Bokeh plot by calling specific glyph functions from the figure object. The glyph functions correspond to different types of glyphs that can be used to represent data. These functions include circle(), line(), rect(), vbar(), hbar(), patch(), and more.

```python
from bokeh.plotting import figure, show

# Create a figure object
p = figure()

# Define the data points
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Add a circle glyph to represent the data points
p.circle(x, y, size=10, color='blue')

# Display the plot
show(p)
```

##Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
Ans -

Axes:

- Title: Set the title of the x-axis and y-axis using the xaxis.axis_label and yaxis.axis_label attributes, respectively.
Tick labels and formatting: Customize the tick labels and formatting of the axes using attributes like xaxis.major_label_text_font_size or yaxis.major_label_text_color.
Axis limits: Adjust the limits of the axes using x_range and y_range attributes of the figure object.
Gridlines: Control the visibility and styling of gridlines using attributes like xgrid.visible and ygrid.grid_line_color.-

- Title:
Set the plot title using the title.text attribute of the figure object.
Customize the appearance of the title using attributes like title.text_font_size or title.text_color.

- Legend:Add a legend to the plot using the legend attribute of the glyph function. For example, p.circle(..., legend_label="Data").
Customize the position and appearance of the legend using the legend attribute of the figure object. For example, p.legend.location = "top_right".



```python
from bokeh.plotting import figure, show

# Create a figure object
p = figure()

# Add glyphs and customize appearance
p.circle(x, y, size=10, color='blue', legend_label="Data")
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.title.text = "My Bokeh Plot"
p.title.text_font_size = "18pt"
p.legend.location = "top_right"

# Display the plot
show(p)
```

##Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
Ans -
- The Bokeh server is a key component of Bokeh that allows you to create and deploy interactive plots and applications that can be updated in real time. It provides the infrastructure to run Bokeh plots as standalone web applications with server-side capabilities.

- With the Bokeh server, you can create plots that respond to user interactions, update dynamically based on changing data, and synchronize across multiple clients. It enables real-time interactivity by maintaining a WebSocket connection between the server and the client, allowing bidirectional communication.

1)Import the necessary modules: Import the required modules from the Bokeh library, including curdoc for accessing the current document and figure for creating the plot.

2)Create the plot: Create the plot as you would normally using the figure() function. Customize the plot appearance, add glyphs, and configure any interactions.

3)Define callback functions: Define the functions that will be executed in response to user interactions or data updates. These functions will update the plot or its data dynamically.

4)Add callbacks to the plot: Attach the callback functions to specific events or interactions using the on_event() or add_periodic_callback() methods of the figure object.

5)Run the Bokeh server: Start the Bokeh server to serve the plot and handle interactions. Use curdoc() to access the current document and add the plot to it.

6)Run the server: Run the Bokeh server to start the application. You can use the Bokeh command-line interface or the bokeh serve command to run the server.

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route("/")
def index():
    # Create the Bokeh plot
    p = figure(...)
    p.circle(...)

    # Generate the plot components
    script, div = components(p)

    # Render the HTML template and pass the plot components
    return render_template("index.html", script=script, div=div)

if __name__ == "__main__":
    app.run()
```
