##Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans -
- Matplotlib is a popular data visualization library for Python. It provides a wide range of tools and functions for creating static, animated, and interactive visualizations in various formats. Matplotlib is widely used in the scientific and data analysis communities for generating high-quality plots and charts.

- Line Plot: A line plot displays data points connected by straight lines. It is commonly used to visualize trends or relationships between variables over a continuous interval.

- Scatter Plot: A scatter plot represents individual data points as markers on a graph. It is useful for displaying the relationship between two numeric variables and identifying patterns or clusters in the data.

- Bar Plot: A bar plot uses rectangular bars to represent data. It is commonly used to compare discrete categories and their associated values, such as comparing sales figures across different products or visualizing survey results.

- Histogram: A histogram is a graphical representation of the distribution of a dataset. It consists of a series of bins, where the height of each bar represents the frequency or count of data points falling within that bin.

- Pie Chart: A pie chart represents data as slices of a circular pie. Each slice corresponds to a specific category and shows the proportion or percentage of that category relative to the whole. Pie charts are commonly used to show composition or relative proportions of different parts in a dataset.



##Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create the scatter plot
plt.scatter(x, y , c ='r')

# Add title and labels
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Display the plot
plt.show()
```

##Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Ans - The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to plot multiple plots (subplots) in a single figure, arranged in rows and columns. This is useful when you want to visualize and compare multiple plots side by side

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Data for line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with subplots
plt.figure(figsize=(12, 10))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, c='b')
plt.title('Line Plot 1')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2,c='r')
plt.title('Line Plot 2')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, c='g')
plt.title('Line Plot 3')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4,c='y')
plt.title('Line Plot 4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

##Q4:Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


Ans -A bar plot, also known as a bar chart or bar graph, is a type of plot that uses rectangular bars to represent categorical data. It displays the values of different categories as bars with lengths proportional to the values they represent. Bar plots are commonly used to compare and visualize data across different categories or groups.

- Comparison: Bar plots make it easy to compare the values of different categories visually. By comparing the heights or lengths of the bars, you can quickly identify which categories have higher or lower values.

- Distribution: Bar plots can show the distribution of data across different categories. They help to identify patterns, trends, or anomalies in the data.

- Ranking: Bar plots can be used to rank categories based on their values. They provide a clear visual representation of the relative sizes or quantities of different categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')

# Display the bar plot
plt.show()
```

##Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Ans -
- A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a concise summary of the distribution of a dataset. It displays key statistical measures, such as the median, quartiles, and potential outliers, in a compact and intuitive manner.

- Box plots are used for the following purposes:

Identifying central tendency: Box plots show the median, which represents the central value of the dataset. The position of the median within the box indicates the skewness of the distribution.

Understanding variability: The interquartile range (IQR), represented by the box, provides insight into the spread or variability of the data. The longer the box, the greater the dispersion in the dataset.

Detecting outliers: Box plots display potential outliers as individual data points beyond the whiskers. Outliers can be valuable in identifying unusual observations or extreme values.

Comparing distributions: Box plots are useful for comparing the distributions of multiple datasets side by side. They allow for quick visual comparison of the central tendency, spread, and skewness across different groups.



```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
# Create a bar plot
plt.figure(figsize=(14, 8))
plt.bar(box1, box2)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')

# Customize x-axis labels
plt.xticks([1, 2], ['Box 1', 'Box 2'])

# Display the box plot
plt.show()
```
